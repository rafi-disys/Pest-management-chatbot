```python
import requests
from IPython.display import HTML
import pandas as pd
```

```python
api_endpoint = 'https://api.cabi.org/FactsheetAdmin/pw/factsheets/IN'
response = requests.get(api_endpoint)
print(response)
```

```python
data = response.json()
print(data)
```

```python
with open('factsheet_dataset2.txt', 'w', encoding='utf-8') as txt_file:
    for factsheet in data['Factsheets']:
        title = factsheet['Title']
        authors = ', '.join([f"{author['FirstName']} {author['MiddleInitial']} {author['Surname']}" for author in factsheet['Authors']])
        editors = ', '.join([editor['FirstName'] for editor in factsheet['Editors']])
        sections = '\n    '.join([section['Value'] for section in factsheet['Sections']])
        scientific_names = ', '.join(factsheet['ScientificNames'])
        images = ', '.join([image['FileName'] for image in factsheet['Images']])

        # Clean HTML tags from sections
        cleaned_sections = []
        for section in factsheet['Sections']:
            section_text = section['Value']
            # section_text = BeautifulSoup(section_text, 'html.parser').get_text()
            # cleaned_sections.append(section_text)

            soup = BeautifulSoup(section_text, 'html.parser')
            # Extract the text content without HTML tags
            text_content = soup.get_text()

            cleaned_sections.append(text_content)
        sections = '\n    '.join(cleaned_sections)
        
        txt_file.write(f"Title: {title}\n")
        txt_file.write(f"Authors: {authors}\n")
        txt_file.write(f"Editors: {editors}\n")
        txt_file.write("Sections:\n")
        txt_file.write(f"    {sections}\n")
        txt_file.write(f"Scientific Names: {scientific_names}\n")
        txt_file.write(f"Images: {images}\n")
        txt_file.write(f"Content:\n{' '.join([section['Value'] for section in factsheet['Sections']])}\n\n")
        txt_file.write("=" * 80 + "\n")  # Add a delimiter

```

```python
from bs4 import BeautifulSoup

# Assuming you have the 'data' dictionary containing your factsheet data

output_content = ""
for factsheet in data['Factsheets']:
    title = factsheet['Title']
    authors = ', '.join([f"{author['FirstName']} {author['MiddleInitial']} {author['Surname']}" for author in factsheet['Authors']])
    editors = ', '.join([editor['FirstName'] for editor in factsheet['Editors']])
    
    cleaned_sections = []
    for section in factsheet['Sections']:
        section_text = section['Value']

        soup = BeautifulSoup(section_text, 'html.parser')
        text_content = soup.get_text()

        cleaned_sections.append(text_content)
    sections = '\n    '.join(cleaned_sections)
    
    scientific_names = ', '.join(factsheet['ScientificNames'])
    images = ', '.join([image['FileName'] for image in factsheet['Images']])

    content = f"Title: {title}\n" \
              f"Authors: {authors}\n" \
              f"Editors: {editors}\n" \
              f"Sections:\n    {sections}\n" \
              f"Scientific Names: {scientific_names}\n" \
              f"Images: {images}\n" \
              f"Content:\n{' '.join([section['Value'] for section in factsheet['Sections']])}\n\n" \
              f"{'=' * 80}\n"

    output_content += content

# Print the generated content to the output area
print(output_content)
```

```python
from bs4 import BeautifulSoup

# Assuming you have the 'data' dictionary containing your factsheet data

output_content = ""
for factsheet in data['Factsheets']:
    title = factsheet['Title']
    authors = ', '.join([f"{author['FirstName']} {author['MiddleInitial']} {author['Surname']}" for author in factsheet['Authors']])
    editors = ', '.join([editor['FirstName'] for editor in factsheet['Editors']])
    sections = []

    for section in factsheet['Sections']:
        section_text = section['Value']

        soup = BeautifulSoup(section_text, 'html.parser')
        text_content = soup.get_text()

        sections.append(text_content)

    formatted_sections = '\n    '.join(sections)
    scientific_names = ', '.join(factsheet['ScientificNames'])
    images = ', '.join([image['FileName'] for image in factsheet['Images']])

    content = f"Title: {title}\n" \
              f"Authors: {authors}\n" \
              f"Editors: {editors}\n" \
              f"Sections:\n    {formatted_sections}\n" \
              f"Scientific Names: {scientific_names}\n" \
              f"Images: {images}\n" \
              f"{'=' * 80}\n"   

    output_content += content
    
    # Print the generated content for each factsheet
    print(content)

# Print the generated content to the output area
print(output_content)
```
